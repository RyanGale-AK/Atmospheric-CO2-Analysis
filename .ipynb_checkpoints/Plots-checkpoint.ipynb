```python
%pylab inline
import matplotlib.pyplot as plt
from pylab import *
```

```python
ls7  = loadtxt('least_squares_7.dat', unpack=True)
ls5  = loadtxt('least_squares_5.dat', unpack=True)
ls3  = loadtxt('least_squares_3.dat', unpack=True)
t = loadtxt('time.dat', unpack=True)
original = loadtxt('original.dat', unpack=True)
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p

plt.plot(t, original,'black')
p.ylabel('$CO2$ ($in$ $ppm$)')
p.xlabel('$time$ ($in$ $months$)')
p.suptitle('$Atmospheric$ $CO_2$')
p.savefig('AtmosphericCO2.png',dpi=1000, bbox_inches='tight')
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p

plt.plot(t, ls7,'red', label='$Least$ $Squares$')
plt.plot(t, original,'black', label='$Original$')
plt.plot(t, ls3,'green', label='$Without$ $Seasonal$ $Variation$')
plt.legend(bbox_to_anchor=(1.01, 1), loc=0, borderaxespad=0.)
p.ylabel('$CO2$ ($in$ $ppm$)')
p.xlabel('$time$ ($in$ $months$)')
p.suptitle('$7$ $Parameter$ $Fit$')
p.subplots_adjust(top=0.90)
p.savefig('7params.png',dpi=1000, bbox_inches='tight')
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p

plt.plot(t, ls5,'red', label='$Least$ $Squares$')
plt.plot(t, original,'black', label='$Original$')
plt.plot(t, ls3,'green', label='$Without$ $Seasonal$ $Variation$')
plt.legend(bbox_to_anchor=(1.01, 1), loc=0, borderaxespad=0.)
p.ylabel('$CO2$ ($in$ $ppm$)')
p.xlabel('$time$ ($in$ $months$)')
p.suptitle('$5$ $Parameter$ $Fit$')
p.subplots_adjust(top=0.90)
p.savefig('5params.png',dpi=1000, bbox_inches='tight')
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p

plt.plot(t, ls7,'red', label='$7$ $parameter$')
plt.plot(t, ls5,'blue', label='$5$ $parameter$')
plt.plot(t, ls3,'green', label='$Without$ $Seasonal$ $Variation$')
plt.legend(bbox_to_anchor=(1.01, 1), loc=0, borderaxespad=0.)
p.ylabel('$CO2$ ($in$ $ppm$)')
p.xlabel('$time$ ($in$ $months$)')
p.suptitle('$7$ $vs$ $5$ $Parameter$ $Fit$')
p.subplots_adjust(top=0.90)
p.savefig('7vs5params.png',dpi=1000, bbox_inches='tight')
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p

residual = []
for index,time in enumerate(t):
    res = ls7[index]-original[index]
    residual.append(res)

plt.plot(t, residual,'black')
p.ylabel('$CO2$ ($in$ $ppm$)')
p.xlabel('$time$ ($in$ $months$)')
p.suptitle('$7$ $Parameter$ $Residual$')
p.savefig('residual7.png',dpi=100, bbox_inches='tight')
```
